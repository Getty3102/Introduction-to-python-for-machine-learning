```python
cols = ['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']

stock = []

for col in cols:
    stock_var = data.groupby(['Area', 'Element'])[col].sum()
    stock.append(stock_var)

stock_df = pd.concat(stock, axis=1).reset_index()
stock_df.columns = ['Area', 'Element'] + cols

madagascar_stock = stock_df[stock_df['Area'] == 'Madagascar']

madagascar_stock
```
